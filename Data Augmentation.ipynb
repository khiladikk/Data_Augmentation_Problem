```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
imggen = ImageDataGenerator(
    rotation_range = 25,
    width_shift_range = 0.20,
    height_shift_range = 0.20,
    shear_range = 0.2,
    zoom_range = 0.50,
    horizontal_flip = 0.30,
    fill_mode = "nearest"
)
```

```python
in_batch= x.reshape((1, *x.shape))

in_batch.shape
```

```python
i = 0
new_batch = []
for out_batch in imggen.flow(in_batch, batch_size=1):
    plt.figure()
    
    
    
    imgplot = plt.imshow(image.img_to_array(out_batch[0]))
    i+= 1 
    
    
#     image.save_img(path=r"C:\Users\kk\dataaug",target_size=(100,100),class_mode= 'binary', file_format= ".jpg", range(50),scale= True **kwargs)
    
    
    if i == 5:
        break
        
    plt.axis("off")
    plt.show()
```
